```python
import numpy as np
```

```python

# ایجاد آرایه از لیست
arr1 = np.array([1, 2, 3, 4, 5])
print("آرایه یک بعدی:", arr1)
```

```python

# آرایه دو بعدی (ماتریس)
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print("آرایه دو بعدی:")
print(arr2d)
```
