```python
class Car:
    # مشخصه کلاس (بین همه اشیاء مشترک)
    wheels = 4
    
    def __init__(self, brand, model, color):
        # مشخصات شیء (برای هر شیء متفاوت)
        self.brand = brand
        self.model = model
        self.color = color
        self.speed = 0
    
    def accelerate(self, amount):
        self.speed += amount
        return f"سرعت به {self.speed} کیلومتر بر ساعت رسید"
    
    def brake(self):
        self.speed = 0
        return "توقف کردید"
    
    def get_info(self):
        return f"{self.brand} {self.model} - رنگ: {self.color}"

# استفاده از کلاس
car1 = Car("تویوتا", "کمری", "نقره‌ای")
car2 = Car("هیوندای", "النترا", "سفید")

print(car1.get_info())  # تویوتا کمری - رنگ: نقره‌ای
print(car1.accelerate(50))  # سرعت به 50 کیلومتر بر ساعت رسید
print(car1.wheels)  # 4  
```
