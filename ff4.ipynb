```python
def aval(a):
    r=list(range(1,int(a/2)+1))
    for i in r:
        if i==1:
            continue
        if a % i != 0 :
            r.remove(i)
        print(r)            
    if len(r) == 1: 
        print("adad is aval")
    else:
        print("adad is morakab")
                    
        
            
```

```python
aval(7)
```
